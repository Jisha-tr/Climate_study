```python
env_df=pd.read_csv("Data/Environment_Temperature_change_E_All_Data_NOFLAG.csv",encoding='ISO-8859-1') 
env_df.head(5)
```

## Selecting the feature

```python
env_df.info()
```

```python
#get only where Element is Temperature change
env_df = env_df[env_df['Element'] == 'Temperature change']
```

```python
#droppping the below col from the table with only element observation has tempreature change
env_df.drop(columns=['Area Code', 'Months Code', 'Element Code','Element', 'Unit'], axis=1, inplace=True)
env_df.head()
```

## Adding Month

```python
#Get values where months is in : 'January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'
env_df = env_df[env_df['Months'].isin(['January', 'February', 'March', 'April', 'May', 'June', 'July',
       'August', 'September', 'October', 'November', 'December'])]
env_df['Months'].unique()
```

```python
#Get the columns dataframe
env_df.columns
```

```python
env_df = pd.melt(env_df, id_vars = ['Area', 'Months'], var_name = 'years', value_name = 'temperature')
env_df.head()
```
